```python
# Importar bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar archivos CSV
precios_hoteles = pd.read_csv('valenciahoteles.csv', encoding='latin-1')
```

```python
precios_hoteles.head(20)
```

```python
# Estadísticas descriptivas del DataFrame resultante
precios_hoteles.describe()
```

```python
# Crear un gráfico de barras con los datos de la columna 'Precio' y 'Nombre'
precios_hoteles.plot.bar(x='Nombre', y='Precio', rot=0)
# Mostrar el gráfico
plt.show()
```

```python
# Crear un gráfico de dispersión con los datos de la columna 'Precio' y 'Valoración'
precios_hoteles.plot.scatter(x='Valoración', y='Precio')

# Mostrar el gráfico
plt.show()
```

```python
# Filtrar los hoteles cuyo precio es mayor a 80
precios_hoteles[precios_hoteles['Precio'] > 80]
```

```python
# Ordenar los hoteles por valoración en orden descendente y obtener los primeros 10 hoteles
precios_hoteles.sort_values(by='Valoración', ascending=False).head(10)
```

```python
# Agrupar los datos por distrito y calcular la suma de los precios por distrito
precios_por_distrito = precios_hoteles.groupby('Distrito')['Precio'].sum()

# Crear un diagrama de sectores con los datos agrupados por distrito
precios_por_distrito.plot.pie(autopct='%.1f%%')

# Mostrar el gráfico
plt.show()
```

```python
# Agrupar los datos por distrito y calcular el promedio de las valoraciones por distrito
valoraciones_por_distrito = precios_hoteles.groupby('Distrito')['Valoración'].mean()

# Crear un diagrama de sectores con los datos agrupados por distrito
valoraciones_por_distrito.plot.pie(autopct='%.1f%%')

# Mostrar el gráfico
plt.show()
```

```python
# Agrupar los datos por distrito y calcular el número de hoteles por distrito
hoteles_por_distrito = precios_hoteles.groupby('Distrito')['Nombre'].count()

# Crear un diagrama de sectores con los datos agrupados por distrito
hoteles_por_distrito.plot.pie(autopct='%.1f%%')

# Mostrar el gráfico
plt.show()
```
