```python
# Crear un gráfico de violín con los datos de la columna 'Precio'
plt.figure(figsize=(10, 6))
sns.violinplot(data=precios_hoteles, y='Precio')

# Añadir etiquetas a los ejes
plt.xlabel('')
plt.ylabel('Precios')

# Añadir un título al gráfico
plt.title('Distribución de precios de los hoteles')

# Agregar etiquetas adicionales
plt.text(0.05, 0.9, f"Mínimo: {precios_hoteles['Precio'].min()}", transform=plt.gca().transAxes)
plt.text(0.05, 0.85, f"Cuartil 25%: {precios_hoteles['Precio'].quantile(0.25)}", transform=plt.gca().transAxes)
plt.text(0.05, 0.8, f"Mediana: {precios_hoteles['Precio'].median()}", transform=plt.gca().transAxes)
plt.text(0.05, 0.75, f"Cuartil 75%: {precios_hoteles['Precio'].quantile(0.75)}", transform=plt.gca().transAxes)
plt.text(0.05, 0.7, f"Máximo: {precios_hoteles['Precio'].max()}", transform=plt.gca().transAxes)

# Mostrar el gráfico
plt.show()
```

```python
# Crear un gráfico de caja (box plot) con los datos de la columna 'Precio'
plt.figure(figsize=(10, 6))
sns.boxplot(data=precios_hoteles, y='Precio')

# Añadir etiquetas a los ejes
plt.xlabel('')
plt.ylabel('Precios')

# Agregar etiquetas adicionales
plt.text(0.05, 0.9, f"Mínimo: {precios_hoteles['Precio'].min()}", transform=plt.gca().transAxes)
plt.text(0.05, 0.85, f"Cuartil 25%: {precios_hoteles['Precio'].quantile(0.25)}", transform=plt.gca().transAxes)
plt.text(0.05, 0.8, f"Mediana: {precios_hoteles['Precio'].median()}", transform=plt.gca().transAxes)
plt.text(0.05, 0.75, f"Cuartil 75%: {precios_hoteles['Precio'].quantile(0.75)}", transform=plt.gca().transAxes)
plt.text(0.05, 0.7, f"Máximo: {precios_hoteles['Precio'].max()}", transform=plt.gca().transAxes)

# Añadir un título al gráfico
plt.title('Distribución de precios de los hoteles')

# Mostrar el gráfico
plt.show()
```

```python
# Crear un gráfico de dispersión con los datos de la columna 'Precio' y 'Valoración', clasificado por distrito
plt.figure(figsize=(8, 6))
sns.scatterplot(data=precios_hoteles, x='Valoración', y='Precio', hue='Distrito')

# Añadir etiquetas a los ejes
plt.xlabel('Valoración')
plt.ylabel('Precio')

# Añadir un título al gráfico
plt.title('Dispersión Precios vs Valoracion, segun el Distrito')

# Mostrar el gráfico
plt.show()
```

```python
# Agrupar los datos por distrito y calcular el número de hoteles por distrito
hoteles_por_distrito = precios_hoteles.groupby('Distrito')['Nombre'].count()

# Ordenar los distritos de mayor a menor
hoteles_por_distrito = hoteles_por_distrito.sort_values(ascending=False)

# Crear un diagrama de sectores con los datos agrupados por distrito
fig, ax = plt.subplots(figsize=(8, 6))
wedges, labels, _ = ax.pie(hoteles_por_distrito, autopct='%.1f%%', wedgeprops=dict(width=0.5))

# Ajustar el aspecto del gráfico
ax.set_aspect('equal')
ax.set_title('Número de hoteles por distrito')

# Agregar enumeración a las etiquetas de los sectores
for i, label in enumerate(labels):
    label.set_text(f'{i+1}: {label.get_text()}')

# Crear una leyenda separada con los distritos ordenados y la enumeración
legend_labels = [f'{i+1}: {label} ({hoteles_por_distrito[label]:.0f} hoteles)' for i, label in enumerate(hoteles_por_distrito.index)]
ax.legend(wedges, legend_labels, title='Distritos', loc='center left', bbox_to_anchor=(1.1, 0.5))

# Mostrar el gráfico
plt.show()
```

```python
# Agrupar los datos por distrito y calcular el promedio de los precios por distrito
precios_por_distrito = precios_hoteles.groupby('Distrito')['Precio'].mean()

# Ordenar los distritos de mayor a menor
precios_por_distrito = precios_por_distrito.sort_values(ascending=False)

# Crear un diagrama de anillo con los datos agrupados por distrito
fig, ax = plt.subplots()
wedges, labels, _ = ax.pie(precios_por_distrito, autopct='%.1f%%', wedgeprops=dict(width=0.4))

# Ajustar el aspecto del gráfico
ax.set_aspect('equal')
ax.set_title('Promedio de precios por distrito')

# Agregar enumeración a las etiquetas de los sectores
for i, label in enumerate(labels):
    label.set_text(f'{i+1}: {label.get_text()}')

# Crear una leyenda separada con los distritos ordenados y la enumeración
legend_labels = [f'{i+1}: {label} (${precios_por_distrito[label]:.1f})' for i, label in enumerate(precios_por_distrito.index)]
ax.legend(wedges, legend_labels, title='Distritos', loc='center left', bbox_to_anchor=(1, 0.5))

# Mostrar el gráfico
plt.show()
```

```python
# Agrupar los datos por distrito y calcular el promedio de las valoraciones por distrito
valoraciones_por_distrito = precios_hoteles.groupby('Distrito')['Valoración'].mean()

# Ordenar los distritos de mayor a menor
valoraciones_por_distrito = valoraciones_por_distrito.sort_values(ascending=False)

# Crear un diagrama de anillo con los datos agrupados por distrito
fig, ax = plt.subplots()
wedges, labels, _ = ax.pie(valoraciones_por_distrito, autopct='%.1f%%', wedgeprops=dict(width=0.4))


# Ajustar el aspecto del gráfico
ax.set_aspect('equal')
ax.set_title('Promedio de Valoraciones por Distrito')

# Agregar enumeración a las etiquetas de los sectores
for i, label in enumerate(labels):
    label.set_text(f'{i+1}: {label.get_text()}')

# Crear una leyenda separada con los distritos ordenados
legend_labels = [f'{i+1}: {label} ({valoraciones_por_distrito[label]:.1f})' for i, label in enumerate(valoraciones_por_distrito.index)]
ax.legend(wedges, legend_labels, title='Distritos', loc='center left', bbox_to_anchor=(1, 0.5))


# Mostrar el gráfico
plt.show()

```
